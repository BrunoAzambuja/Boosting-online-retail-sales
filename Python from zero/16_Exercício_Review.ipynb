```python
# abrir o arquivo de excel e gerar dataframe
planilha = pd.read_excel('/content/gdrive/My Drive/controle_da_empresa.xlsx', None)
df = planilha['Vendas']
```

```python
# vendo os tipos de dados por colunas
df.dtypes
```

```python
# plotar gráfico de linhas do valores de vendas (Janeiro/2020)
df.plot(x = 'Data da Venda', y = 'Total de Vendas')

plt.title("Valores de Vendas (Janeiro/2020)")
plt.xlabel("Data da Venda")
plt.ylabel("Valor Total do Dia")

plt.show()
```

```python
# soma de requisições por itens
itens = df['Item'].value_counts()
itens
```

```python
# plotar gráfico de barras da soma de requisições por itens (Janeiro/2020)
itens.plot(x = 'itens', y = 'Quantidade', kind = 'bar')

plt.title("Requisições por Itens (Janeiro/2020)")
plt.xlabel("Itens")
plt.ylabel("Quantidade de venda")

plt.show()
```

```python
# plotar gráfico de histograma da quantidade de ocorrências por faixa de valor (Janeiro/2020)
plt.hist('Total de Vendas', data = df, bins = 10)
plt.title("Histograma de Vendas")
plt.xlabel("Faixa de Valor")
plt.ylabel("Frequência de Ocorrências")
plt.show
```
